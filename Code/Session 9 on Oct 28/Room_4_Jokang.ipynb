You have been given **IMDB sentiment analysis Dataset**. This is a dataset for binary sentiment classification containing. We provide a set of 40,000 highly polar movie reviews for training, and 10,000 for testing. Raw text and already processed bag of words formats are provided.

Your task is to implement a **1 hidden Layer Neural Network with 100 hidden neurons** for classifying a review as a positive review or a negative review.


### Task 1: Implement Steps 3 to 7 below:
* Step 1: Load Data from Keras.datasets (Already Implemented) 
* Step 2: Vectorize input data with 10,000 words (Already Implemented)
* Step 3: Build a sequential Neural Netowrk Keras model with 1 hidden dense layer with 100 hidden neurons (https://www.tensorflow.org/api_docs/python/tf/keras/Sequential)
* Step 4: Initialize SGD optimizer (https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/SGD)
* Step 5: Compile Model (https://www.tensorflow.org/api_docs/python/tf/keras/losses/CategoricalCrossentropy)
* Step 6: Fit Model on Training Data
* Step 7: Evaluate Model on Test Data (Find Test Accuracy)

https://www.tensorflow.org/api_docs/python/tf/keras/Model: Compile, Fit, Evaluate Model

```
# Step 1 (Already Implemented)
# https://keras.io/api/datasets/imdb/#load_data-function
import tensorflow as tf
(x_train, y_train), (x_test, y_test)  = tf.keras.datasets.imdb.load_data(num_words=10000)

# Step 2 (Already Implemented)
import numpy as np
data = np.concatenate((x_train, x_test), axis=0)
targets = np.concatenate((y_train, y_test), axis=0)

# Visualization
index = tf.keras.datasets.imdb.get_word_index()
reverse_index = dict([(value, key) for (key, value) in index.items()]) 
decoded = " ".join( [reverse_index.get(i - 3, "#") for i in data[0]] )
print(decoded) 

def vectorize(sequences, dimension = 10000):
  results = np.zeros((len(sequences), dimension))
  for i, sequence in enumerate(sequences):
    results[i, sequence] = 1
  return results
 
data = vectorize(data)
targets = np.array(targets).astype("float32")

x_train = data[10000:]
y_train = targets[10000:]
x_test = data[:10000]
y_test = targets[:10000]
```

```
np.unique(y_train)
```

```
#Step 3 (To be implemented)
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(100,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))

#Step 4 (To be implemented)
sgd = tf.keras.optimizers.SGD(learning_rate=0.1, momentum = 0.01)
#Step 5 (To be implemented)
loss1 = tf.keras.losses.BinaryCrossentropy()
nn_model.compile(optimizer = sgd, loss = loss1, metrics = ['accuracy'])
#Step 6 (To be implemented)
nn_model.fit(x_train, y_train,epochs = 10)
#Step 7 (To be implemented)
_, accuracy = nn_model.evaluate(x_test, y_test)
print ('Test Accuracy: %2f' % (accuracy*100))
```

### Task 2: Add more dense layers to build a deeper Neural Network (NN) model and tune hyper-parameters to further increase testing accuracy. Report best test accuracy with a deep NN with tuned hyper-parameters.

```
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(100,input_dim=x_train.shape[1], activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(30,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(20,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(20,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(20,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(10,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(10,activation = 'relu'))
#nn_model.add(tf.keras.layers.Dense(10,activation = 'relu'))
nn_model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))

#Step 4 (To be implemented)
sgd = tf.keras.optimizers.SGD(learning_rate=0.05, momentum = 0.01)
#Step 5 (To be implemented)
loss1 = tf.keras.losses.BinaryCrossentropy()
nn_model.compile(optimizer = sgd, loss = loss1, metrics = ['accuracy'])
#Step 6 (To be implemented)

nn_model.fit(x_train, y_train,epochs = 10, validation_data=(x_test,y_test))

#Step 7 (To be implemented)
_, accuracy = nn_model.evaluate(x_test, y_test)
print ('Test Accuracy: %2f' % (accuracy*100))
```
