You have been given a **Reuters Dataset**. This is a dataset of 11,228 newswires from Reuters, labeled over 46 topics. This was originally generated by parsing and preprocessing the classic Reuters-21578 dataset. 

Each newswire is encoded as a list of word indexes (integers). For convenience, words are indexed by overall frequency in the dataset, so that for instance the integer "3" encodes the 3rd most frequent word in the data. This allows for quick filtering operations such as: "only consider the top 10,000 most common words". 

Your task is to implement a **1 hidden Layer Neural Network with 100 hidden neurons** for classifying a newswire from Reuters as one of 46 topics.

### Task 1: Implement Steps 3 to 8 below:
* Step 1: Load Data from Keras.datasets (Already Implemented) 
* Step 2: Vectorize input data with 10,000 words (Already Implemented)
* Step 3: One hot encoding of training and testing labels (https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical)
* Step 4: Build a sequential Neural Netowrk Keras model with 1 hidden dense layer with 100 hidden neurons (https://www.tensorflow.org/api_docs/python/tf/keras/Sequential)
* Step 5: Initialize SGD optimizer (https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/SGD)
* Step 6: Compile Model (https://www.tensorflow.org/api_docs/python/tf/keras/losses/CategoricalCrossentropy)
* Step 7: Fit Model on Training Data
* Step 8: Evaluate Model on Test Data (Find Test Accuracy)

https://www.tensorflow.org/api_docs/python/tf/keras/Model: Compile, Fit, Evaluate Model

```
# Step 1 (Already Implemented)
# https://keras.io/api/datasets/reuters/#load_data-function
import tensorflow as tf
(x_train, y_train), (x_test, y_test)  = tf.keras.datasets.reuters.load_data(num_words=10000)

# Step 2 (Already Implemented)
import numpy as np
data = np.concatenate((x_train, x_test), axis=0)
targets = np.concatenate((y_train, y_test), axis=0)

index = tf.keras.datasets.reuters.get_word_index()
reverse_index = dict([(value, key) for (key, value) in index.items()]) 
decoded = " ".join( [reverse_index.get(i - 3, "#") for i in data[0]] )
print(decoded)

def vectorize(sequences, dimension = 10000):
  results = np.zeros((len(sequences), dimension))
  for i, sequence in enumerate(sequences):
    results[i, sequence] = 1
  return results
 
data = vectorize(data)

targets = np.array(targets).astype("float32")

x_train = data[2000:]
y_train = targets[2000:]
x_test = data[:2000]
y_test = targets[:2000]
```

```
# Step 3 (To be implemented)
y_train_encoded = tf.keras.utils.to_categorical(y_train)
y_test_encoded = tf.keras.utils.to_categorical(y_test)

# Step 4 (To be implemented)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1000, input_shape=(10000,), activation='sigmoid'))
model.add(tf.keras.layers.Dense(500, activation='sigmoid'))
model.add(tf.keras.layers.Dense(200, activation='sigmoid'))
model.add(tf.keras.layers.Dense(46, activation='softmax'))

# Step 5 (To be implemented)
opt = tf.keras.optimizers.SGD(learning_rate=.5)

# Step 6 (To be implemented)
cce = tf.keras.losses.CategoricalCrossentropy()
model.compile(opt, cce, metrics=['accuracy'])

# Step 7 (To be implemented)
model.fit(x_train, y_train_encoded, validation_data=(x_test, y_test_encoded), epochs=100)

# Step 8 (To be implemented)
model.evaluate(x_test, y_test_encoded)
model.summary()
```
