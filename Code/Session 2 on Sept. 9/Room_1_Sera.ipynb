### **Q2:** In the following code cell implement the following:
* Step 1: Import the dataset (tae.csv) using Pandas Dataframe (Step 1 Implemented already)
* Step 2: Partition your dataset into training testing and validation using sklearns train_test_split library
* Step 3: Split the features and target labels into seperate variables
* Step 4: Scale the features using sklearns min max scaling function
* Step 5: If there is an off the shelf library of the ML algorithm, you can try and input the data to the library.

```
# Step 1 already implemented
import pandas as pd
import io
import requests
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
import numpy

url="https://raw.githubusercontent.com/Mihir2/BreakoutSessionDataset/master/tae.csv"
s = requests.get(url).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))
results_train = []
results_test = []

for i in range(50):
  Train, Test= train_test_split(data, test_size=0.1, shuffle=True)
  data

  Train_y = Train.iloc[:,[0]]
  Train_x = Train.iloc[:,1:]

  Test_y = Test.iloc[:,[0]]
  Test_x = Test.iloc[:,1:]

  scaler = MinMaxScaler() 
  scaler.fit(Train_x)
  Train_x_scaled = scaler.transform(Train_x)
  Test_x_scaled = scaler.transform(Test_x)

  reg = LogisticRegression().fit(Train_x_scaled, Train_y.values.ravel())
  results_train.append(reg.score(Train_x_scaled, Train_y.values.ravel()))
  results_test.append(reg.score(Test_x_scaled, Test_y.values.ravel()))

print(numpy.mean(results_train))
print(numpy.mean(results_test))
```
