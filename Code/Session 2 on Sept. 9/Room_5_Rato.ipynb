### **Q2:** In the following code cell implement the following:
* Step 1: Import the dataset (student-cse4-574.csv) using Pandas Dataframe (Step 1 Implemented already)
* Step 2: Map the strings in each feature column to numbers
* Step 3: Partition your dataset into training testing and validation using sklearns train_test_split library
* Step 4: Split the features and target labels into seperate variables
* Step 5: Scale the features using sklearns min max scaling function
* Step 6: If there is an off the shelf library of the ML algorithm, you can try and input the data to the library.

```
# Step 1 already implemented
import pandas as pd
import io
import requests

from sklearn.preprocessing import MinMaxScaler

url="https://raw.githubusercontent.com/Mihir2/BreakoutSessionDataset/master/student-cse4-574.csv"
s = requests.get(url).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))

diag_map = {'M':1, 'F':0}
data['sex'] = data['sex'].map(diag_map)

diag_map = {'U':1, 'R':0}
data['address'] = data['address'].map(diag_map)

diag_map = {'GT3':1, 'LE3':0}
data['famsize'] = data['famsize'].map(diag_map)

diag_map = {'A':1, 'T':0}
data['Pstatus'] = data['Pstatus'].map(diag_map)

diag_map = {'teacher':0, 'health':1, 'services':2, 'at_home':3, 'other':4}
data['Mjob'] = data['Mjob'].map(diag_map)

diag_map = {'teacher':0, 'health':1, 'services':2, 'at_home':3, 'other':4}
data['Fjob'] = data['Fjob'].map(diag_map)

diag_map = {'home':0, 'reputation':1, 'course':2, 'other':3}
data['reason'] = data['reason'].map(diag_map)

diag_map = {'mother':0, 'father':1, 'other':2}
data['guardian'] = data['guardian'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['schoolsup'] = data['schoolsup'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['famsup'] = data['famsup'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['paid'] = data['paid'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['activities'] = data['activities'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['nursery'] = data['nursery'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['higher'] = data['higher'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['internet'] = data['internet'].map(diag_map)

diag_map = {'yes':1, 'no':0}
data['romantic'] = data['romantic'].map(diag_map)

from sklearn.model_selection import train_test_split
y=data['Grade ']
X=data.drop(columns=['Grade '])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler = MinMaxScaler() 
scaler.fit(X_train)
Train_x_scaled = scaler.transform(X_train)
Test_x_scaled = scaler.transform(X_test)

## TA Answer: Implement a Neural Network using sklearn
from sklearn.neural_network import MLPClassifier

# Training MLP
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(Train_x_scaled,y_train)

# Testing MLP
clf.predict(Test_x_scaled)
```
