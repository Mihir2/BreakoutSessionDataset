### **Q2:** In the following code cell implement the following:
* Step 1: Import the dataset (USOpen-men-2013.csv) using Pandas Dataframe (Step 1 Implemented already)
* Step 2: Partition your dataset into training testing and validation using sklearns train_test_split library
* Step 3: Split the features and target labels into seperate variables
* Step 4: Scale the features using sklearns min max scaling function
* Step 5: If there is an off the shelf library of the ML algorithm, you can try and input the data to the library.

```
# Step 1 already implemented
import pandas as pd
import io
import requests
from sklearn.model_selection import train_test_split  #step 3
from sklearn.preprocessing import MinMaxScaler        #step 4
from sklearn.linear_model import LogisticRegression   #step 5
url="https://raw.githubusercontent.com/Mihir2/BreakoutSessionDataset/master/USOpen-men-2013.csv"
s = requests.get(url).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))
data
X = data.loc[:, data.columns != 'Result'] #step 2
Y = data['Result']                        #step 2
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42) #step 3
mmscaler = MinMaxScaler()                 #step 4
X_mmscaler_train = mmscaler.fit_transform(X_train)#step 4
X_mmscaler_test = mmscaler.transform(X_test)      #step 4
logmodel = LogisticRegression()           #step 5
logmodel.fit(X_mmscaler_train,Y_train)    #step 5
```

### **Q3:**  How would you evaluate your trained model?

Answer: 

```
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
predictions = logmodel.predict(X_mmscaler_test)
print(confusion_matrix(Y_test, predictions))
print(classification_report(Y_test, predictions)) #here goes
```
